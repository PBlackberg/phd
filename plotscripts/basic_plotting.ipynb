```python
#packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import seaborn as sns
import matplotlib.gridspec as gridspec
import matplotlib.colors as colors

import matplotlib as mpl
```

```python
# basic lineplot

x_data= np.arange(0,101,10)
y_data= x_data**2
noise= np.random.randint(-10,10,11)
y_data= np.add(y_data,noise)

fig = plt.figure(1, figsize=(5,5))
plt.plot(x_data,y_data, marker='o', markersize=7.5, mfc= 'white', linestyle='--',label='$y=x^2$')
plt.plot(x_data,y_data+500, marker='s', markersize=7.5, mfc= 'white', linestyle='-',label='$y=x^2+500$')

plt.xlabel('x-variable')
plt.ylabel('y-label')
plt.legend(loc='lower right')
plt.show()


#print(len(x_data))
```

```python
# basic barplot

rings= {'elves':3, 'men':9, 'dwarves':7, 'sauron':1}
fig = plt.figure(1, figsize=(5,5))
lists = sorted(rings.items(), key=lambda item:item[1])
x,y = zip(*lists)

plt.bar(x,y)
#plt.barh(x,y)

# pie chart
#explode = (0,0.1,0,0)
#plt.pie(y, labels=x, explode=explode)
```

```python
# basic scatterplot
x_data= np.arange(0,50,1)
y_data= np.arange(0,50,1)
colors= np.arange(0,50,1)
area= np.random.rand(50)**2*1000

fig = plt.figure(1, figsize=(5,5))
plt.scatter(x_data, y_data, c=colors, s=area, alpha=0.8)
```

```python
# basic histogram
failure_data= np.random.normal(size=3000)*200

fig = plt.figure(1, figsize=(5,5))
plt.hist(failure_data, bins=30)
```

```python
# multiple panels


#mpl.rcParams['pdf.fonttype']=42
#mpl.rcParams['font.size']=14

filename= r'test.csv'
df= pd.read_csv(filename)

df= df.iloc[::10,:]

x90=df['X 90 Kohm'].dropna()
y90=df['Y 90 Kohm'].dropna()
x90_clear=df['X 90 Kohm-clear'].dropna()
y90_clear=df['Y 90 Kohm-clear'].dropna()

x60=df['X 60 Kohm'].dropna()
y60=df['Y 60 Kohm'].dropna()
x60_clear=df['X 60 Kohm-clear'].dropna()
y60_clear=df['Y 60 Kohm-clear'].dropna()

x30=df['X 30 Kohm'].dropna()
y30=df['Y 30 Kohm'].dropna()
x30_clear=df['X 30 Kohm-clear'].dropna()
y30_clear=df['Y 30 Kohm-clear'].dropna()

xmin= 400
xmax= 800
ymin= 60
ymax= 120

colors= sns.color_palette('rocket',6)

# figure
fig = plt.figure(1, figsize=(5,5))
gs= gridspec.GridSpec(4,4)
gs.update(wspace=0.25,hspace=0.25)


# first plot
xtr_subplot= fig.add_subplot(gs[0:4,0:2])

plt.plot(x90,y90,linestyle='none',marker='^',label='90',color=colors[0],mfc='w',markersize=8)
plt.plot(x60,y60,linestyle='none',marker='^',label='60',color=colors[1],mfc='w',markersize=8)
plt.plot(x30,y30,linestyle='none',marker='^',label='30',color=colors[2],mfc='w',markersize=8)

xticks= np.arange(300,701,200)
yticks= np.arange(60,121,20)
plt.minorticks_on()
plt.tick_params(direction='in',which='minor',length=5, bottom='true', top='true',left='true',right='true')
plt.tick_params(direction='in',which='major',length=10, bottom='true', top='true',left='true',right='true')
plt.xticks(xticks)
plt.yticks(yticks)

plt.legend()

plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

plt.ylabel('transparancy %')
plt.text(650,52,'wavelength (nm)') # based on the y-axis of the first plot




# second plot
xtr_subplot= fig.add_subplot(gs[0:2,2:4])

plt.plot(x90_clear,y90_clear,linestyle='none',marker='^',label='90 clear',color=colors[3],mfc='w',markersize=8)
plt.plot(x60_clear,y60_clear,linestyle='none',marker='^',label='60 clear',color=colors[4],mfc='w',markersize=8)
plt.plot(x30_clear,y30_clear,linestyle='none',marker='^',label='30 clear',color=colors[5],mfc='w',markersize=8)

xticks= np.arange(100,800,200)
#yticks= np.arange(60,121,20)
plt.minorticks_on()
plt.tick_params(direction='in',which='minor',length=5, bottom='true', top='true',left='true',right='true')
plt.tick_params(direction='in',which='major',length=10, bottom='true', top='true',left='true',right='true')
plt.tick_params(labelbottom=False, labeltop=False, labelright=True, labelleft=False)

plt.xticks(xticks)
plt.yticks(yticks)

plt.legend()

plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)


# third plot
xtr_subplot= fig.add_subplot(gs[2:4,2:4])

plt.plot(x90_clear,y90_clear,linestyle='none',marker='^',label='90 clear',color=colors[3],mfc='w',markersize=8)
plt.plot(x60_clear,y60_clear,linestyle='none',marker='^',label='60 clear',color=colors[4],mfc='w',markersize=8)
plt.plot(x30_clear,y30_clear,linestyle='none',marker='^',label='30 clear',color=colors[5],mfc='w',markersize=8)

xticks= np.arange(100,800,200)
#yticks= np.arange(60,121,20)
plt.minorticks_on()
plt.tick_params(direction='in',which='minor',length=5, bottom='true', top='true',left='true',right='true')
plt.tick_params(direction='in',which='major',length=10, bottom='true', top='true',left='true',right='true')
plt.tick_params(labelbottom=True, labeltop=False, labelright=True, labelleft=False)

plt.xticks(xticks)
plt.yticks(yticks)

plt.legend()

plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)







```
